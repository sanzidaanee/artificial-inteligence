```python
# Define a function to detect outliers using IQR
def detect_outliers_iqr(df):
    outliers = {}
    for col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    return outliers
```

```python
# Display outliers
for col, outlier_data in outliers.items():
    print(f"\nOutliers in {col}:")
    print(outlier_data)
```

```python
# Remove outliers using IQR method for numeric columns only
def remove_outliers_iqr(data):
    numeric_cols = data.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    return data

df_no_outliers = remove_outliers_iqr(df)
```

```python
# Display the DataFrame without outliers
print("\nDataFrame after Removing Outliers:")
print(df_no_outliers.head())
```

**Data Visualization**

```python
def plot_continuous_distribution(df = None, column: str = None, height: int = 8):
  _ = sns.displot(df, x=column, kde=True, height=height, aspect=height/5).set(title=f'Distribution of {column}');
```

```python
plot_continuous_distribution(df, 'unit_price')
plot_continuous_distribution(df, 'quantity')
plot_continuous_distribution(df, 'total')
```

```python
def get_unique_values(df, column):
  num_unique_values = len(df[column].unique())
  value_counts = df[column].value_counts()
  print(f"Column: {column} has {num_unique_values} unique values\n")
  print(value_counts)
```

```python
get_unique_values(df, 'category')
get_unique_values(df, 'customer_type')
get_unique_values(df, 'payment_type')
```

```python
def plot_categorical_distribution(data: pd.DataFrame = None, column: str = None, height: int = 8, aspect: int = 2):
  _ = sns.catplot(data=data, x=column, kind='count', height=height, aspect=aspect).set(title=f'Distribution of {column}');
```

```python
plot_categorical_distribution(df, 'category')
plot_categorical_distribution(df, 'customer_type')
plot_categorical_distribution(df, 'payment_type')
```
